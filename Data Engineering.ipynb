```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
confirmed = pd.read_csv("time_series_covid19_confirmed_global.csv")
```

```python
confirmed
```

```python
confirmed.drop("Province/State", axis = 1, inplace = True)
```

```python
confirmed = confirmed.melt(id_vars = ["Country/Region", "Lat", "Long"], 
                          var_name = "Date",
                          value_name = "ConfirmedCase")
```

```python
confirmed
```

```python
confirmed.to_csv("confirmed_case.csv", index = False)
```
